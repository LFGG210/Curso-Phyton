JSON es un formato para guardar y transmitir datos. Es muy APIs, archivos de configuracion y almacenamiento de informacion
(se parace a un diccionario de Python, peor es solo texto)

Para que sirve?
- Sirve para convertir objetos de python
- texto JSON y tambien texto JSON -> objetos de python

Esto se llama:
- Serializar: Convertir un objeto de python en texto JSON
- Deserializar: lo opuesto

```python
# Funciones principales
#1) json.dums(obj): Convierte diccionario u objeto en Python
import json
obj = {"nombre": "Ana"
,"edad": 25}
print(type(obj))
s = json.dumps(obj, ensure_ascii = False)
print(s)
print(type(s))
```

```python
# Json.loads(texto_json): Convierte un string JSON en un diccionario python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
#Al venir de una paguina/persona/archivo desconocid@

va = '''
Pueden contener:
Datos malinformados
Informacion inesperada
Campos faltantes
Datos demasiado grandes
variables incorrectas que rompan tu programa
'''
print(va)
```

# Numpy
provee el tipo ndarray, operaciones vectorizadas, brodcasting, utilidades de algebra lineal y funciones matematicas en arrays.

Es la base sobre la que se apoyan muchas librerias cinetificas en python

### Nota importante: el arreglo de Numpy no es iagual a la lista de python esto no puede teneer diferentes tipos de datos en el mismo

```python
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) #array (2,3)
v = np.array([10,20,30]) # array (3)
print(A+v)
print("------------------")
prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba+prueba2)
```

```python
# mas funciones de numpy
import numpy as np
z = np.zeros((3,3))
print(z)
o = np.ones((2,4))
print(o)
r = np.random.random((3,3)) #matriz
print(r)
```

```python
#arngos numericos
nums = np.arange(0,10,2)#saca los valores entre 0 y 10 de 2 en 2
print(nums)
x = np.linspace(0,1,5) #saca los 5 valores espaciodos estre 0 y 1
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
```

```python
print(a.dot(b))#productos entre dos matrices
c  = np.array([[1,2,3],[4,5,6]])
d = np.array([[10,20],[30,40],[50,60]])
print(c.dot(d))
```

```python
data = np.array([1,2,3,4,5])
print(data.max())
print(data.min())
print(data.sum())
print(data.mean())
print(data.std())
```

```python
A = np.array([1,6,3,9,2])
print(A[A > 5])
```
