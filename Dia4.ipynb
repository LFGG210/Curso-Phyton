## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adiós, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return f"{saludo}\n{despedida}"

nombre_usuario = input("Ingrese su nombre: ")
mensaje = mensaje_completo(nombre_usuario)
print(mensaje)
```

```python
# Supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ['vainilla',"chocolate","fresa"]
  print(f"Sabor disponible, {sabores}")
  sabor = input("Elige un sabor:").strip().lower()
  while sabor not in sabores:
    sabor = input("Elige un sabor:").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve?: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve?(Ingresa un valor valido): "))
  return cantidad

#Funcion Madre(Llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de de sabor {sabor}")
total_pedidos = int(input("Cuantos clientes atenderas?: "))
for _ in range(total_pedidos):
  pedir_helado()
```

# Funciones lambda
Una funcion lambda es una funcion acortada para escribir funcion en forma:

nombreFuncion = lambda parametros: accion

la diferencia entre las funciones lambda y las normales es que son mas limitadas

```python
sumar = lambda x,y : x+y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1 - desc)), precio * desc)
nuevo_precio = descuentoLambda(0.15,500)
print(f"Eprecio a pagar es: {nuevo_precio[0]}, el descuento que se hizo fue de: {nuevo_precio[1]}")
```

```python
from ast import While
#Calculadora -> suma, resta, producto, division y potencia
def suma(a,b):
  return a+b
def resta(a,b):
  return a-b
def producto(a,b):
    return a*b
def division(a,b):
  if b == 0:
    return("No se puede dividir entre cero")
  return a/b
def potencia(a,b):
  return a**b

def menu():
  print("\n---------MI CALCULADORA------------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. Division")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("\nSeleccione una opcion:"))
    if op < 1 or op > 6:
      print("Opcion invalida")

  except ValueError:
    print("Ingrese un valor numerico")
  if op == 6:
    print("\nSaliendo...")
    break

  try:
    a = float(input("\nIngrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
  except ValueError:
    print("Ingrese un valor numerico")
  if op == 1:
    print(f"\nEl resultado de la suma es: {suma(a,b)}")
  elif op == 2:
    print(f"\nEl resultado de la resta es: {resta(a,b)}")
  elif op == 3:
    print(f"\nEl resultado del producto es: {producto(a,b)}")
  elif op == 4:
    print(f"\nEl resultado de la division es: {division(a,b)}")
  elif op == 5:
    print(f"\nEl resultado de la potencia es: {potencia(a,b)}")
  else:
    print("Opcion invalida")
```
