Operaciones matematicas

```python
# Cuerto nivel de importancia
# Suma ordinaria
print(3 + 4)
variable1,variable2 = 20,40
print(variable1 + variable2)

# Resta Ordinaria
print(3 - 4)
v1 = 10.4
v2 = 15
Resultado = v2 - v1
print(Resultado)
```

```python
# Tercer nivel de inportancia
# Multiplicacion ordinaria
print("Multiplicacion ordinaria")
M1, M2 = 2, 3.14
print(M1 * M2)
# Division Ordinaria
print("---------------------")
print("Division ordinaria")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M5)
print(M5 / M6)
#DIVISION DE PISO
print("Division de piso")
print(M1 // M2)
print(M3 // M5)
print(M5 // M6)
```

```python
# Potencias
print("potencias")
print(2 ** 3)
p1,p2 = 7,2
print(p1 ** p2)
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
print(A)
```
