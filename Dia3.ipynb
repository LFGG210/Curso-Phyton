# Listas
Las listas son colecciones ordenadas que podemos modificar y almacenar alementos de distintos tipos de datos

```python
lista_de_compras = ["manzanas","pan","leche","huevos","mantequilla","queso","jamon"]
print(lista_de_compras)

print(lista_de_compras[0])
print(lista_de_compras[1])
print(lista_de_compras[2])
print(lista_de_compras[3])
```

```python
#Agregar un elemento a la lista .append
lista_de_compras.append("sopa")
lista_de_compras
```

```python
#Agregar un elemento pero en un lugar en especifico
lista_de_compras.insert(1,"platano")
lista_de_compras
```

```python
# Para eliminar un elemento de una lista existe el remove y el pop
lista_de_compras.remove("platano")
lista_de_compras
```

```python
lista_de_compras.pop() # Sin nada dentro del paratesis elimina el ultimo elemento
lista_de_compras.pop(7) # o puedes agarrar la posicion que quieras (pero iniciando desde 0)
#Nota te retorna el numero eliminado
```

```python
# Podemos recorrer el arreglo por medio de un for
for elementos in lista_de_compras:
  print(elementos)
```

```python
# len() saca el tamano de la lista
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i + 1} - {lista_de_compras[i]}")
```

```python
promedios = [55.6,97.5,23.6,97.5,72.4,100]
print(promedios)
promedios.sort() #sort sirve para odenar la lista de manera ascendente
print(promedios)
#ordenar una listad e manera descendente
promedios.sort(reverse=True)
print(promedios)

#Para investigar mas buscar metodos de ordenamiento
```

```python
# Podemos ingresar al ultimo elemento de la lista
print(promedios[-1])
```

```python
# Encontrar el mayor y el menor de una lista
print(max(promedios))
print(min(promedios))
```

# Sets
Los sets son colecciones SIN orden y que no permiten elementos duplicados

```python
nombres = {"Angel","Eliud","Diana","Juan","Juan"}
print(nombres)
print(type(nombres))
```

```python
distintos_datos = {"Angel", 1, "Diana", "Juan", "Juan", 4.8}
print(distintos_datos)
```

```python
#Add
nombres.add("Leonardo")
nombres
```

```python
print(nombres[2]) #Va a marcar error porque los sets no tienen orden
```

```python
# Union
matutino = {20322,22183,123123,53234}
vespertino = {20333,11111,2222,33333}
toltal_matriculas = matutino.union(vespertino)
print(toltal_matriculas)
toltal_matriculas = matutino | vespertino
print(toltal_matriculas)
```

```python
#Interseccion
ca = {1234,121,646,12414,73434,124124,675234}
al = {1234,111,222,333,444}
#interseccion con .intersection()
ca_al = ca.intersection(al)
print(ca_al)
#interseccion con &
ca_al = ca & al
print(ca_al)
```

```python
#Diferencia
solo_calculo = ca.difference(al)
print(solo_calculo)
# Diferencia usando -
solo_calculo = ca -al
print(solo_calculo)


solo_algebra = al.difference(ca)
print(solo_algebra)
```

```python
nombre = {"Angel","Eliud","Diana","Juan","Juan"}
#remove y pop
nombre.remove("Juan") #Puedes eliminar el que tu quieras
print(nombre)
nombre.pop() # Elimina un elemento de  manera aleatoria
print(nombre)
```

```python
# Eliminar todo el SET
nombres.clear()
nombres
```

# Diccionarios
Los diccionarios son colecciones no ordenadas de pares clave: valor

```python
alumno = {
    #clave     valor
"nombre": "Fernando Garcia",
"edad": 21,
"Carrera": "IMC",
"semestre": 1,
"promedio": 88,
"calificaciones": [100,94,94,97,100]
}
```

```python
print("El alumno", alumno["nombre"], "esta estudiando", alumno["Carrera"], "en el semestre", alumno["semestre"])
```

```python
#Agregar al diccionario
alumno["area_preferida"] = "Diseno CAD"
print(alumno)
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for aspecto, valor in alumno.items(): #devuelve una vista de todos los pares calve:valor del diccionario
  print(aspecto, valor)
```

```python
# El metodo .clear tambien se puede utilizar aqui
#Tenemos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))

alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
# El metodo .values() ssirve para obtener los valores del diccionario, tambien se puede usar en la interaccion
alumno_v = alumno.values()
print(alumno_v)
print(type(alumno_v))
```

```python
if "Carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")

if "Dinero" in alumno:
  print("Alumno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
#.copy()puedes copiar todo el diccionario en una nueva variable
# update -> Sirve para cambiar los values (.update())
alumno.update({"semestre": 6})
alumno
```
