Condicional Elif
Ahora hagamos un programa en el que preguntamos la calificacion del 0 al 100 y queremos saber su equivalemte en el sistema de evaluacion estadounidense, basemonos en esto:
A+ = 100
A = 90-99
B = 80-89
C = 70-79
D = 60-69
F < 60

```python
cal = int(input("Ingresa tu calificacion: "))

if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: # al ya no haber una condcion el else cubro todo lo demas
  print("F")
```

```python
cal = int(input("Ingresa tu calificacion: "))
if  cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else: # al ya no haber una condcion el else cubro todo lo demas
    print("F")
else:
  print("Error, la calificacion tiene que estar entre 0 y 100")
```

```python
# El modulo (%)
# El mopdulo es un operador matematico que devuelve el residuo de una division
# 10 % 3 da 1,porque 10 dividido entre 3 cabe 3 veces y sobra 1
# 7 % 2 da 1, porque el 7 no es divisible de forma exacta entre el 2

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es multiplo de 3 o 5")

# Saber si el numero es par o impar
num = 6
if not num % 2 == 0:
  print(f"El numero {num} es impar")
else:
  print(f"El numero {num} es par")
```

Ciclo for
Los llamados ciclos for nos ayudan a realizar acciones durante una cantidad determinada de veces

```python
# Imprimir los numeros del 0 al 4
for i in range(5): #Toma 5 como el numero final pero inicia siempre en 0
  print(i)
```

```python
for num in range(2,11):
   print(num)
```

While

```python
suma = 0
numeros = int(input("Ingresa la cantidad de numeros a sumar o 0 para salir: "))

while numeros != 0: # != es equivalente a distinto de....
  suma += numeros
  numeros = int(input("Ingresa la cantidad de numeros a sumar o 0 para salir: "))
print(f"La suma de los numeros es: {suma}")
```

```python
# Usar el while como for
numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```
